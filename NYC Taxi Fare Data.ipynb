```python
df_crd_only = df[df[' payment_type'] == 'CRD']
df_crd_sample = df_crd_only.loc[sample(df_crd_only.index, 10000)]
df_crd_sample.head()
```

```python
tip_threshold = 0.12
df_crd_sample['tipped'] = (df_crd_sample[' tip_amount'] / df_crd_sample[' fare_amount'] >= tip_threshold).astype(int)

good_tips = df_crd_sample[df_crd_sample['tipped']==1]
bad_tips = df_crd_sample[df_crd_sample['tipped']==0]

tips_pct = df_crd_sample[' tip_amount'] / df_crd_sample[' fare_amount']
tips_pct[tips_pct < 2].hist(bins=30)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.scatter(x=good_tips['pickup_latitude'], y=good_tips['pickup_longitude'],color='b',marker='o', s=0.1, alpha=0.5)
ax1.scatter(x=bad_tips['pickup_latitude'], y=bad_tips['pickup_longitude'],color='r',marker='o', s=0.1, alpha=0.5)
plt.xlim(40.63,40.83)
plt.ylim(-74.05,-73.76)

ax2 = fig.add_subplot(212)
ax2.scatter(x=good_tips['dropoff_latitude'], y=good_tips['dropoff_longitude'],color='b',marker='o', s=0.1, alpha=0.5)
ax2.scatter(x=bad_tips['dropoff_latitude'], y=bad_tips['dropoff_longitude'],color='r',marker='o', s=0.1, alpha=0.5)
plt.xlim(40.63,40.83)
plt.ylim(-74.05,-73.76)
```

```python
# Drop feats of no interest
df_crd_sample.drop(['medallion','hack_license', ' hack_license', 'store_and_fwd_flag', ' payment_type', ' vendor_id', ' pickup_datetime'],inplace=True,axis=1)
df_crd_sample.info()
```

```python
vendor_id_dummies = pd.get_dummies(df_crd_sample['vendor_id'])
df_crd_sample = pd.merge(df_crd_sample, vendor_id_dummies, how='outer', left_index=True, right_index=True)
df_crd_sample.drop(['vendor_id'],inplace=True,axis=1)

rate_code_dummies = pd.get_dummies(df_crd_sample['rate_code'], prefix='rate_code')
df_crd_sample = pd.merge(df_crd_sample, rate_code_dummies, how='outer', left_index=True, right_index=True)
del df_crd_sample['rate_code']

# Drop feats dependent on outcome
df_crd_sample.drop([' tip_amount',' total_amount'], inplace=True,axis=1)
```

```python
df_crd_sample['pickup_weekday'] = pd.to_datetime(df_crd_sample['pickup_datetime']).apply(lambda x: x.weekday())
df_crd_sample['dropoff_weekday'] = pd.to_datetime(df_crd_sample['dropoff_datetime']).apply(lambda x: x.weekday())
df_crd_sample['pickup_minute'] = pd.to_datetime(df_crd_sample['pickup_datetime']).apply(lambda x: x.hour * 60 + x.minute)
df_crd_sample['dropoff_minute'] = pd.to_datetime(df_crd_sample['dropoff_datetime']).apply(lambda x: x.hour * 60 + x.minute)
del df_crd_sample['pickup_datetime']
del df_crd_sample['dropoff_datetime']

df_crd_sample.head()
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split( \
    df_crd_sample.drop(['tipped'], axis=1), \
    df_crd_sample['tipped'], test_size=0.4, random_state=0)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

tuned_parameters = [{'alpha' : [0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]}]
clf = GridSearchCV(SGDClassifier(loss="modified_huber"), tuned_parameters, cv=5, scoring="recall")
clf.fit(X_train, y_train)

print("Best parameters set found on development set:")
print()
print(clf.best_estimator_)
print()
print("Grid scores on development set:")
print()
for params, mean_score, scores in clf.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r"
          % (mean_score, scores.std() / 2, params))
print()

print("Detailed classification report:")
print()
print("The model is trained on the full development set.")
print("The scores are computed on the full evaluation set.")
print()
y_true, y_pred = y_test, clf.predict(X_test)
print(classification_report(y_true, y_pred))
print()
```

```python
pd.Series(y_pred).value_counts()
```

```python
fpr, tpr, thr = roc_curve(y_true, y_pred)
auc = roc_auc_score(y_true, y_pred)
print "AUC: " + str(auc)

plt.plot(fpr,tpr) #e
plt.plot(fpr,fpr) #e
plt.xlabel("False positive rate")   #e
plt.ylabel("True positive rate") #e
```
